# Random Walk

````python
import numpy as np
import pandas as pd
import scipy
import random
import matplotlib.pylab as plt
````

````python
x = [0]
for j in range(1000):
    x.append(x[j] + np.random.normal())
````

````python
df=pd.Series(x)
plt.figure(figsize=(12,6));
df.plot()
plt.show()
````

# rho=0

# rho=0.5

````python
rho = 0.5
x = [0]
for j in range(1000):
    x.append( rho*x[j] + np.random.normal())
    
df=pd.Series(x)
plt.figure(figsize=(12,6));
df.plot()
plt.show()  
````
