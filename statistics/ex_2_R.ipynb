# Regresja liniowa

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
```

# Przykład
Proszę narysować wykres funkcji 
$$
f(x)=(2x-1)(x^2-2)(x-2)+3
$$

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_tr, y_tr, '--k');
plt.show()
```

```python
f_lr_1 = lambda x: 0.1*x +0.1
f_lr_2 = lambda x: 0.1*pow(x,2) + 0.1*x + 0.1 
f_lr_3 = lambda x: 0.1*pow(x,3) + 0.1*pow(x,2) + 0.1*x + 0.1 
f_lr_4 = lambda x: 0.1*pow(x,4) + 0.1*pow(x,3) + 0.1*pow(x,2) + 0.1*x+ 0.1 
f_lr_5 = lambda x: 0.1*pow(x,5) + 0.1*pow(x,4) + 0.1*pow(x,3) + 0.1*pow(x,2) + 0.1*x + 0.1 
```
